## **MUMBAI WEATHER ANALYSIS**

```python
mumbai_df = df[df['city'] == 'Mumbai']
monthly = mumbai_df.groupby("month").agg({
    "avg_temperature_C": "mean",
    "rainfall_mm": "mean"
}).reset_index()
```

```python
# "Walter leith climate diagram"-
#The Walter-Lieth climate diagram is used to visualize annual temperature (red line) and precipitation (blue line) patterns, determining humid/arid periods at a glance.
# It assists in characterizing biomes, agricultural planning, and analyzing climate-vegetation relationships by mapping ecological moisture conditions
import matplotlib.pyplot as plt

# Create figure
fig, ax1 = plt.subplots()

# Line plot for temperature
ax1.plot(monthly["month"], monthly["avg_temperature_C"], marker='o', label="Temperature (°C)")
ax1.set_xlabel("Month")
ax1.set_ylabel("Temperature (°C)")

# Bar plot for rainfall (scaled)
ax2 = ax1.twinx()  # second y-axis
ax2.bar(monthly["month"],monthly["rainfall_mm"] / 2,alpha=0.3, label="Rainfall (mm/2)")
ax2.set_ylabel("Rainfall (mm/2)")

# Title
plt.title("Walter-Lieth Diagram - Mumbai")

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
bars, bar_labels = ax2.get_legend_handles_labels()
ax1.legend(lines + bars, labels + bar_labels)

plt.show()
```

```python
# Seasonal Subseries Plot- A seasonal subseries plot is a specialized time series visualization that groups data by season (e.g., all Januaries, all Februaries)
# to reveal seasonal patterns, trends within specific seasons, and changes over time.

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

colors = plt.cm.tab20(np.linspace(0, 1, mumbai_df["month"].nunique()))

for i, (month, data) in enumerate(mumbai_df.groupby("month")):


    data = data.sort_values(by="day")

    plt.plot(
        data["day"],
        data["avg_temperature_C"],
        color=colors[i],
        linewidth=1.5,
        alpha=0.7,
        label=month
    )

plt.xlabel("Day")
plt.ylabel("avg_temperature_C")
plt.title("Seasonal Subseries Plot (Temperature - Mumbai)")

plt.grid(True, linestyle='dotted', linewidth=0.7, alpha=0.7)
plt.legend(title="month", bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap- A correlation heatmap is a 2D graphical representation of a data matrix that
#  uses colors to visualize the strength and direction of relationships (correlation coefficients) between multiple numerical variables
import seaborn as sns
import matplotlib.pyplot as plt

cols = [
    'max_temperature_C',
    'min_temperature_C',
    'avg_temperature_C',
    'humidity_%',
    'rainfall_mm',
    'cloud_cover_%',
    'wind_speed_km_h',
    'avg_AQI',
    'PM2_5',
    'PM10',
    'avg_visibility_morning_km',
    'avg_visibility_afternoon_km',
    'avg_visibility_night_km'
]

# Correlation matrix
corr = mumbai_df[cols].corr()

# Plot
plt.figure(figsize=(12, 10))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="plasma",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.8}
)

plt.title("Meteorological Correlation Heatmap - Mumbai")

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Calender heatmap- A calendar heatmap is a data visualization tool
#that maps time-series data—such as daily, weekly, or monthly measurements—onto a grid layout resembling a calendar.
import seaborn as sns
import matplotlib.pyplot as plt

# Create pivot table (Month vs Day)
pivot = mumbai_df.pivot_table(
    values='avg_temperature_C',
    index='month',
    columns='day'
)

plt.figure(figsize=(12, 6))

sns.heatmap(
    pivot,
    cmap="plasma",
    linewidths=0.3,
    linecolor='white',
    cbar_kws={"label": "Temp (°C)"}
)

# Labels and title
plt.xlabel("Day")
plt.ylabel("Month")
plt.title("Temperature Calendar Heatmap - Mumbai")

# Cleaner ticks
plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Multiple box plots- A multibox plot (more commonly referred to as multiple box plots or side-by-side box plots)
#is a graphical visualization that displays several box-and-whisker plots together in one graph.
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))

sns.boxplot(
    data=mumbai_df,
    x="month",
    y="max_temperature_C",
    palette="viridis",
    linewidth=1.2
)

sns.stripplot(
    data=mumbai_df,
    x="month",
    y="max_temperature_C",
    color='black',
    alpha=0.3,
    size=3
)

plt.title("Monthly Temperature Distribution - Mumbai")
plt.xlabel("Month")
plt.ylabel("Max Temperature (°C)")

plt.tight_layout()
plt.show()
```

```python
# Jointplot (Hexbin Density)-Shows relationship and density between two continuous variables
#                             Better than scatter when points overlap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

sns.histplot(
    data=mumbai_df,
    x="avg_temperature_C",
    y="humidity_%",
    bins=20,
    cbar=True
)

plt.title("Density Plot: Temperature vs Humidity")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Humidity (%)")

plt.tight_layout()
plt.show()
```

```python
# Seasonal decomposition - Seasonal decomposition separates the time series into
#                          trend, seasonal, and residual components to analyze underlying patterns in temperature data.
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Create date column
mumbai_df['date'] = pd.to_datetime(mumbai_df[['year', 'month', 'day']])
mumbai_df = mumbai_df.sort_values('date')

# Time series
ts = mumbai_df.set_index("date")["avg_temperature_C"]

# Decompose
result = seasonal_decompose(ts, model='additive', period=30)

# Create subplots
fig, axes = plt.subplots(4, 1, figsize=(12, 8), sharex=True)

# Original
axes[0].plot(ts)
axes[0].set_title("Original")

# Trend
axes[1].plot(result.trend)
axes[1].set_title("Trend")

# Seasonal
axes[2].plot(result.seasonal)
axes[2].set_title("Seasonal")

# Residual
axes[3].plot(result.resid)
axes[3].set_title("Residual")

# Overall title
fig.suptitle("Seasonal Decomposition - Mumbai")

for ax in axes:
    ax.grid(True, linestyle='dotted', linewidth=0.7, alpha=0.7)

plt.tight_layout()
plt.show()
```
